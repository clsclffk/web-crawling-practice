```python
from bs4 import BeautifulSoup
import requests

url = 'https://www.moviechart.co.kr/rank/realtime/index/image'

response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# 영화 제목
# #content > div.wArea.space > div.movieBox > ul > li:nth-child(1) > div > div.movie-title > h3 > a

movie_list = soup.select('div.movieBox div.movie-title a')
for i in movie_list:
    print(i.text)
```

```python
# 이미지
# #content > div.wArea.space > div.movieBox > ul > li:nth-child(1) > a > img

image_list = soup.select('div.movieBox a img')
for i in image_list:
    print(i['src'])
```

```python
for i in image_list:
    img_response = requests.get('https://www.moviechart.co.kr'+i['src'])
    print(img_response.content)
```

```python
import os

img_dir = "images"

if not os.path.exists(img_dir):
    os.makedirs(img_dir)
```

```python
full_image_dir = os.path.join(os.getcwd(), img_dir)
full_image_dir
```

```python
import re

pattern = re.compile(r'[\\/:"*?<>|]')
```

```python
from PIL import Image
from io import BytesIO


for movie, movie_image in zip(movie_list, image_list):
    img_src = movie_image.get("src")
    img_name = pattern.sub("", movie.text)
    img_response = requests.get('https://www.moviechart.co.kr'+img_src)
    image = Image.open(BytesIO(img_response.content))
    image.save(os.path.join(img_dir, img_name+".jpg"))
```
