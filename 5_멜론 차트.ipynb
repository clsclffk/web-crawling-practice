```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.melon.com/chart/index.htm'
header = {'User-Agent': 'Mozila/5.0 (Windows NT 6.3; Trident/7.0; rv:11.0) like Gecko'}

response = requests.get(url, headers=header)
soup = BeautifulSoup(response.text, 'html.parser')
print(response.text)
```

```python
# #lst50 > td:nth-child(6) > div > div > div.ellipsis.rank01 > span > a
titles = soup.select("div.ellipsis.rank01 a")
print(titles)
```

```python
titles_list = [title.text for title in titles]
titles_list
```

```python
# #lst50 > td:nth-child(6) > div > div > div.ellipsis.rank02 > a:nth-child(1)
singers = soup.select("div.ellipsis.rank02 > a")
singers
```

```python
singers_list = [singer.text for singer in singers]
singers_list
```
